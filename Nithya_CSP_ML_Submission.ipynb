## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#  Remove duplicate rows (if any)
data.drop_duplicates(inplace=True)

#  Handle missing values
# For numerical columns ‚Üí fill with median
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
data[num_cols] = data[num_cols].fillna(data[num_cols].median())

# For categorical columns ‚Üí fill with mode (most common value)
cat_cols = data.select_dtypes(include=['object']).columns
for col in cat_cols:
    data[col] = data[col].fillna(data[col].mode()[0])

#  Convert column names to lowercase (optional for consistency)
data.columns = data.columns.str.lower()

#  Strip extra spaces from text columns (cleaning)
data = data.applymap(lambda x: x.strip() if isinstance(x, str) else x)

#  Verify if dataset is now clean
print("‚úÖ Data Cleaning Completed Successfully!\n")
print("Remaining Missing Values:\n", data.isnull().sum().sum())
print("\nDuplicate Rows Remaining:", data.duplicated().sum())
````

### What all manipulations have you done and insights you found?

Data Manipulations Performed:

1. Removed duplicate rows to make sure each record is unique and accurate.
2. Filled missing values:
   - For number columns, replaced missing values with the median.
   - For text/categorical columns, replaced missing values with the most frequent value (mode).
3. Cleaned column names and removed extra spaces from text entries for better consistency.
4. Checked again to confirm there are no missing or duplicate values left.

Insights After Cleaning:

1. The dataset is now completely clean and ready for analysis.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Here are three good examples from this dataset:

H‚ÇÅ: There is a significant difference in CSAT scores between different agent shifts (Morning, Afternoon, Night).

H‚ÇÇ: There is a significant relationship between item price and CSAT score.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H‚ÇÄ): There is no significant difference in CSAT scores among different agent shifts.
Alternate Hypothesis (H‚ÇÅ): There is a significant difference in CSAT scores among different agent shifts.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy import stats

# Drop missing values
df1 = data.dropna(subset=['agent shift', 'csat score'])

# Group CSAT scores by shift
groups = [group['csat score'].values for _, group in df1.groupby('agent shift')]

# Perform One-Way ANOVA test
f_stat, p_val = stats.f_oneway(*groups)
print("F-Statistic:", f_stat)
print("P-Value:", p_val)

if p_val < 0.05:
    print(" Reject Null Hypothesis: Shift affects CSAT scores.")
else:
    print(" Fail to Reject Null Hypothesis: Shift does not affect CSAT scores.")
````

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA Test

##### Why did you choose the specific statistical test?

Because we are comparing more than two groups (different shifts) to see if their average CSAT scores differ significantly.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H‚ÇÄ): There is no significant relationship between item price and CSAT score.
Alternate Hypothesis (H‚ÇÅ): There is a significant relationship between item price and CSAT score..

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Drop missing values
df2 = data.dropna(subset=['item_price', 'csat score'])

# Pearson correlation test
corr, p_val = stats.pearsonr(df2['item_price'], df2['csat score'])
print("Correlation Coefficient:", corr)
print("P-Value:", p_val)

if p_val < 0.05:
    print(" Reject Null Hypothesis: There is a significant relationship.")
else:
    print(" Fail to Reject Null Hypothesis: No significant relationship found.")
````

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test

##### Why did you choose the specific statistical test?

 Because both variables (item price and CSAT score) are numerical, and we want to measure the strength and direction of their linear relationship.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H‚ÇÄ): There is no significant relationship between customer city and agent shift.
Alternate Hypothesis (H‚ÇÅ): There is a significant relationship between customer city and agent shift

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import chi2_contingency

# Create contingency table for categorical variables
contingency_table = pd.crosstab(data['customer_city'], data['agent shift'])

# Perform Chi-Square Test
chi2, p_val, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("P-Value:", p_val)
print("Degrees of Freedom:", dof)

if p_val < 0.05:
    print(" Reject Null Hypothesis: Customer city and agent shift are related.")
else:
    print(" Fail to Reject Null Hypothesis: No relationship between customer city and agent shift.")
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Define a function to remove outliers using the IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # Clip values (keeps data within bounds)
    df[column] = np.where(df[column] < lower, lower,
                          np.where(df[column] > upper, upper, df[column]))
    return df

# Apply to numerical columns
num_cols = ['item_price', 'connected_handling_time', 'csat score']
for col in num_cols:
    data = remove_outliers_iqr(data, col)

# Verify outlier handling
data[num_cols].describe()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

I used the IQR (Interquartile Range) method to detect and treat outliers in numeric columns.
 Instead of removing outliers, I capped them within acceptable limits to prevent data loss.
 This method is simple, effective, and keeps extreme values from affecting model performance.

````python
# Visualizing evaluation Metric Score chart
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# Calculate Metrics
accuracy = accuracy_score(y_test, y_pred_xgb)
precision = precision_score(y_test, y_pred_xgb, average='weighted')
recall = recall_score(y_test, y_pred_xgb, average='weighted')
f1 = f1_score(y_test, y_pred_xgb, average='weighted')

# Print Scores
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Visualization
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
values = [accuracy, precision, recall, f1]

plt.figure(figsize=(6,4))
plt.bar(metrics, values)
plt.xlabel("Evaluation Metrics")
plt.ylabel("Score")
plt.title("XGBoost Model Performance (Before Tuning)")
plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from skopt import BayesSearchCV
from skopt.space import Integer, Real
from xgboost import XGBClassifier

# Define parameter space for Bayesian Optimization
param_space = {
    'n_estimators': Integer(50, 200),
    'max_depth': Integer(3, 10),
    'learning_rate': Real(0.01, 0.2),
    'subsample': Real(0.6, 1.0),
    'colsample_bytree': Real(0.6, 1.0)
}

xgb_model_bo = XGBClassifier(random_state=42)

# Bayesian Optimization Search
bayes_search = BayesSearchCV(
    estimator=xgb_model_bo,
    search_spaces=param_space,
    n_iter=20,
    cv=3,
    scoring='f1_weighted',
    n_jobs=-1,
    random_state=42
)

# Fit the Algorithm
bayes_search.fit(X_train, y_train)

# Best Tuned Model
best_xgb = bayes_search.best_estimator_

# Predict on the model
y_pred_xgb_tuned = best_xgb.predict(X_test)

print("Best Hyperparameters Found:", bayes_search.best_params_)
````
